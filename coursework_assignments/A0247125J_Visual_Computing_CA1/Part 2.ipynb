# Part 2: SIFT Features and Descriptors

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import numpy.linalg as lin

from skimage.transform import warp
```

i) The image is read and keypoints and descriptors are obtained using the OpenCV SIFT_create() package. The keypoints obtained are then utilized to plot the keypoints on the image.
ii) The radial line in the plot provides the orientation of the keypoint.

```python
im1=plt.imread('im01.jpg')
img=im1.copy()
sift1 = cv2.SIFT_create()
gray_im1= cv2.cvtColor(im1,cv2.COLOR_BGR2GRAY)

kp1, des1 = sift1.detectAndCompute(im1,None)
img=cv2.drawKeypoints(im1,kp1,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img)
```
